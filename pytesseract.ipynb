```python
import cv2 
import pytesseract
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
# بارگذاری تصویر
img = cv2.imread('finalplateimage.jpg')
img=cv2.medianBlur(img, 5)
e,img=cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
cv2.imshow("d",img)
cv2.waitKey(0)
# اضافه کردن تنظیمات سفارشی
# شناسایی متن در تصویر با استفاده از Pytesseract
text = pytesseract.image_to_string(img,config='--psm 6')

# چاپ متن شناسایی شده
print(text)

```

```python
testbox = pytesseract.image_to_boxes(img,config='--psm 6')
print(testbox)
```

```python
testbox=testbox.split("\n")
print(testbox)
```

```python
imgrectangle=cv2.imread('finalplateimage.jpg')
w,h,c=imgrectangle.shape
for T in testbox:
    finalBox=T.split(" ")
    if finalBox[0]:
        x1=int(finalBox[1])
        y1=int(finalBox[2])
        x2=int(finalBox[3])
        y2=int(finalBox[4])
        cv2.rectangle(imgrectangle,(x1,y1),(x2,y2),(255,0,0),5)
plt.imshow(imgrectangle)   
```

```python
imgtext=cv2.imread('amir text.png')
testdata = pytesseract.image_to_data(imgtext, output_type=pytesseract.Output.DICT)
print(testdata)
```

```python
imgtextrectangle=cv2.imread('amir text.png')
for index,text in enumerate(testdata['text']):
             if testdata['text'][index]:
                x1=int(testdata['left'][index])
                y1=int(testdata['top'][index])
                x2=x1+int(testdata['width'][index])
                y2=y1+int(testdata['height'][index])
                cv2.rectangle(imgtextrectangle,(x1,y1),(x2,y2),(255,0,0),1)
plt.imshow(imgtextrectangle)               
```
